```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
def format_func(value, tick_number):
    return '0' if value == 0 else f'{value/1e7:.0f}천만원'
```

```python
# 계약일자 열을 datetime 형식으로 변환
df['계약일자'] = pd.to_datetime(df['계약일자'])

#df4 = df[(df['대장금액(원)'] >= 1) & df['계약일자'].dt.month.between(7, 12)].sort_values(['대장금액(원)', '계약일자'],ascending=[False, True]).head(50)

# 계약일자의 월을 범주형 데이터로 변환
df['월'] = df['계약일자'].dt.month_name()

# 월 기준으로 정렬
order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

# 그래프 생성
colors = ['red', 'gray', 'green', 'lightblue', 'purple', 'pink', 'yellow', 'lightgreen','darkblue', 'orange', 'cyan', 'magenta']
ax = sns.barplot(data=df, x='월', y='대장금액(원)', palette=colors, order=order)
ax.yaxis.set_major_formatter(ticker.FuncFormatter(format_func))

plt.xlabel('계약일자(월)')
plt.ylabel('월별 총 대장금액')
plt.title('계약일자(월) 별 총 대장금액')
plt.xticks(rotation = 90)
plt.show()
```

```python
# 계약일자 열을 datetime 형식으로 변환
df['계약일자'] = pd.to_datetime(df['계약일자'])

#df4 = df[(df['대장금액(원)'] >= 1) & df['계약일자'].dt.month.between(7, 12)].sort_values(['대장금액(원)', '계약일자'],ascending=[False, True]).head(50)

# 계약일자의 월을 범주형 데이터로 변환
df['월'] = df['계약일자'].dt.month_name()

# 월 기준으로 정렬
order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

# 데이터프레임을 월 기준으로 정렬
df_sorted = df.sort_values('월', key=lambda x: pd.Categorical(x, categories=order, ordered=True))

# 그래프 생성
ax = sns.lineplot(data=df_sorted, x='월', y='대장금액(원)', color='black')
ax.yaxis.set_major_formatter(ticker.FuncFormatter(format_func))


plt.xlabel('계약일자(월)')
plt.ylabel('월별 총 대장금액')
plt.title('계약일자(월) 별 총 대장금액 추이')
plt.xticks(rotation=90)
plt.show()
```
